```python
## Data Cleaning
```

```python
df.drop(columns='index', inplace=True)
```

```python
df.head()
```

```python
### Remove Unnecessary Column

The `Unnamed: 22` column contained only `False` and missing values and did not provide any meaningful business information. Therefore, it was removed from the dataset.
```

```python
df['Unnamed: 22'].value_counts(dropna=False)
```

```python
df.drop(columns='Unnamed: 22', inplace=True)
```

```python
df.columns
```

```python
### Standardize Column Names

To improve readability and maintain consistency across Python, SQL, and Power BI, all column names were converted to lowercase snake_case by removing extra spaces and replacing spaces and hyphens with underscores.
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

```python
df.columns
```

```python
## Handling Missing Values
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
### Investigate `fulfilled_by`

The `fulfilled_by` column contains a large number of missing values. Before deciding whether to remove or retain the column, the available values are examined.
```

```python
df['fulfilled_by'].value_counts(dropna=False)
```

```python
### Remove `fulfilled_by`

The `fulfilled_by` column contained only one meaningful value (`Easy Ship`) and a large number of missing values. Since it does not contribute significantly to the sales analysis, the column was removed.
```

```python
df.drop(columns='fulfilled_by', inplace=True)
```

```python
df.columns
```

```python
### Investigate `promotion_ids`

The `promotion_ids` column contains a large number of missing values. The values are examined to determine whether the column is useful for analysis.
```

```python
df['promotion_ids'].value_counts(dropna=False)
```

```python
### Remove `promotion_ids`

The `promotion_ids` column contained promotion campaign information and a large number of missing values. Since promotion performance is outside the scope of this sales analysis project, the column was removed.
```

```python
df.drop(columns='promotion_ids', inplace=True)
```

```python
### Handle Missing Values in `currency`

The `currency` column contained missing values only for cancelled orders. Since all valid records used INR, the missing values were replaced with `INR`.
```

```python
df['currency'].fillna('INR', inplace=True)
```

```python
df['currency'] = df['currency'].fillna('INR')
```

```python
df['currency'].isnull().sum()
```

```python
df['courier_status'].value_counts(dropna=False)
```

```python
### Missing Values in `courier_status`

The missing values in the `courier_status` column were retained because they represent orders for which courier information was not available. These missing values may correspond to cancelled or unprocessed orders and therefore represent valid business scenarios rather than data quality issues.
```

```python
### Convert Date Format

The `date` column was converted to datetime format to enable time-based analysis.
```

```python
df['date'].head(10)
```

```python
### Handle Missing Shipping Information

Only 33 records contained missing shipping details. Since these represent a very small proportion of the dataset, they were removed.
```

```python
df.dropna(
    subset=['ship_city', 'ship_state', 'ship_postal_code', 'ship_country'],
    inplace=True
)
```

```python
df.isnull().sum()
```

```python
## Export Clean Dataset

The cleaned dataset is exported for SQL analysis and Power BI visualization.
```

```python
df.to_csv("amazon_sales_cleaned.csv", index=False)
```
